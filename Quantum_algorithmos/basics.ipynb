```python
import os
import numpy as np
import matplotlib.pyplot as plt
from qiskit.visualization import plot_bloch_vector

# Carpeta donde se guardarán las imágenes
output_dir = "bloch_examples"
os.makedirs(output_dir, exist_ok=True)


def bloch_vector(alpha: complex, beta: complex) -> np.ndarray:
    """
    Calcula el vector de Bloch de un estado alpha|0> + beta|1>.

    Para un estado normalizado:
        x = 2 Re(conj(alpha) beta)
        y = 2 Im(conj(alpha) beta)
        z = |alpha|^2 - |beta|^2
    """
    norm = np.sqrt(abs(alpha)**2 + abs(beta)**2)
    if norm == 0:
        raise ValueError("El estado no puede ser el vector cero.")

    alpha = alpha / norm
    beta = beta / norm

    x = 2 * np.real(np.conjugate(alpha) * beta)
    y = 2 * np.imag(np.conjugate(alpha) * beta)
    z = abs(alpha)**2 - abs(beta)**2

    return np.array([x, y, z], dtype=float)


sqrt2 = np.sqrt(2)

states = {
    "ket_0": {
        "title": r"$|0\rangle$",
        "alpha": 1,
        "beta": 0,
    },
    "ket_1": {
        "title": r"$|1\rangle$",
        "alpha": 0,
        "beta": 1,
    },
    "ket_plus": {
        "title": r"$|+\rangle=(|0\rangle+|1\rangle)/\sqrt{2}$",
        "alpha": 1 / sqrt2,
        "beta": 1 / sqrt2,
    },
    "ket_minus": {
        "title": r"$|-\rangle=(|0\rangle-|1\rangle)/\sqrt{2}$",
        "alpha": 1 / sqrt2,
        "beta": -1 / sqrt2,
    },
    "ket_plus_i": {
        "title": r"$|+i\rangle=(|0\rangle+i|1\rangle)/\sqrt{2}$",
        "alpha": 1 / sqrt2,
        "beta": 1j / sqrt2,
    },
    "ket_minus_i": {
        "title": r"$|-i\rangle=(|0\rangle-i|1\rangle)/\sqrt{2}$",
        "alpha": 1 / sqrt2,
        "beta": -1j / sqrt2,
    },
    "global_phase_plus": {
        "title": r"$e^{i\pi/3}|+\rangle$",
        "alpha": np.exp(1j * np.pi / 3) / sqrt2,
        "beta": np.exp(1j * np.pi / 3) / sqrt2,
    },
}

for filename, data in states.items():
    vector = bloch_vector(data["alpha"], data["beta"])

    fig = plot_bloch_vector(
        vector,
        title=data["title"],
        figsize=(5, 5),
        font_size=14
    )

    path = os.path.join(output_dir, f"{filename}.png")
    fig.savefig(path, dpi=300, bbox_inches="tight")
    plt.close(fig)

    print(f"{filename}: Bloch vector = {vector}, saved in {path}")
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from qiskit.visualization import plot_bloch_vector

output_dir = "bloch_examples"
os.makedirs(output_dir, exist_ok=True)

sqrt2 = np.sqrt(2)


def bloch_vector(alpha: complex, beta: complex) -> np.ndarray:
    norm = np.sqrt(abs(alpha)**2 + abs(beta)**2)
    alpha = alpha / norm
    beta = beta / norm

    return np.array([
        2 * np.real(np.conjugate(alpha) * beta),
        2 * np.imag(np.conjugate(alpha) * beta),
        abs(alpha)**2 - abs(beta)**2
    ], dtype=float)


examples = [
    (r"$|0\rangle$", 1, 0),
    (r"$|1\rangle$", 0, 1),
    (r"$|+\rangle$", 1 / sqrt2, 1 / sqrt2),
    (r"$|-\rangle$", 1 / sqrt2, -1 / sqrt2),
    (r"$|+i\rangle$", 1 / sqrt2, 1j / sqrt2),
    (r"$|-i\rangle$", 1 / sqrt2, -1j / sqrt2),
]

for title, alpha, beta in examples:
    vector = bloch_vector(alpha, beta)
    fig = plot_bloch_vector(vector, title=title, figsize=(4, 4), font_size=12)
    plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector

# ============================================================
# Carpeta de salida
# ============================================================
output_dir = "bloch_pauli_figures"
os.makedirs(output_dir, exist_ok=True)

# ============================================================
# Estados base
# ============================================================
ket0 = Statevector([1, 0])
ket1 = Statevector([0, 1])
ket_plus = Statevector([1/np.sqrt(2), 1/np.sqrt(2)])
ket_minus = Statevector([1/np.sqrt(2), -1/np.sqrt(2)])

# ============================================================
# Matrices de Pauli
# ============================================================
X = np.array([[0, 1],
              [1, 0]], dtype=complex)

Y = np.array([[0, -1j],
              [1j, 0]], dtype=complex)

Z = np.array([[1, 0],
              [0, -1]], dtype=complex)

# ============================================================
# Función para guardar una esfera de Bloch
# ============================================================
def save_bloch(statevector, filename, title):
    fig = plot_bloch_multivector(statevector)
    fig.suptitle(title, fontsize=14)
    fig.savefig(os.path.join(output_dir, filename), bbox_inches="tight", dpi=200)
    plt.close(fig)

# ============================================================
# X sobre |0>
# ============================================================
save_bloch(ket0, "bloch_x_before.png", r"Estado inicial $|0\rangle$")
state_x_after = ket0.evolve(X)
save_bloch(state_x_after, "bloch_x_after.png", r"Estado final $X|0\rangle = |1\rangle$")

# ============================================================
# Y sobre |0>
# ============================================================
save_bloch(ket0, "bloch_y_before.png", r"Estado inicial $|0\rangle$")
state_y_after = ket0.evolve(Y)
save_bloch(state_y_after, "bloch_y_after.png", r"Estado final $Y|0\rangle = i|1\rangle$")

# ============================================================
# Z sobre |+>
# ============================================================
save_bloch(ket_plus, "bloch_z_before.png", r"Estado inicial $|+\rangle$")
state_z_after = ket_plus.evolve(Z)
save_bloch(state_z_after, "bloch_z_after.png", r"Estado final $Z|+\rangle = |-\rangle$")

print(f"Imágenes guardadas en la carpeta: {output_dir}")
print("Archivos generados:")
for f in sorted(os.listdir(output_dir)):
    print(" -", f)

```
